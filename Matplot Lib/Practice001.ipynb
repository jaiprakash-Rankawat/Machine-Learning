```python
import numpy as np
arr1 = np.arange(1,7)
print(arr1)
```

```python
arr2 = 2*arr1
print(arr2)
```

```python
arr3 = np.array([7,15,8,23,41,4])
arr3
```

```python
from matplotlib import pyplot as plt
plt.plot(arr1,arr3)
plt.plot(arr1,arr2,color="Red",linestyle=":",linewidth=3)
plt.title("Graph")
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(arr1,arr3)
plt.subplot(1,2,2)
plt.plot(arr1,arr2,color="Red",linestyle=":",linewidth=3)
plt.title("Graph")
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(arr1,arr3)
plt.subplot(2,1,2)
plt.plot(arr1,arr2,color="Red",linestyle=":",linewidth=3)
plt.title("Graph")
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.show()
```
