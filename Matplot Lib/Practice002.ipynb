```python
# bar graph 
name = ["Ram","Sita","Radha","Krishna"]
Marks = [30,86,100,60]
from matplotlib import pyplot as plt
plt.bar(name,Marks)
plt.title("Student Marks")
plt.xlabel("Student")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
# horizontal bar graph (barh)
plt.barh(name,Marks,color="Red")
plt.title("Student Marks",color="red")
plt.ylabel("Student")
plt.xlabel("Marks")
plt.grid(True)
plt.show()
```

```python
# scatter plot
year = [2010,2012,2014,2016,2018,2020,2022,2024]
marks = [358,315,20,156,46,85,200,16]
plt.plot(year,marks)
plt.show()
```

```python
plt.scatter(year,marks)
plt.show()
```

```python
plt.scatter(year,marks,marker="*",color="red", s=50)
var = [50,100,150,200,250,300,350,400]
plt.scatter(year,var)
plt.show()
```

```python
# converting into sub plot 
plt.subplot(1,2,1)
plt.scatter(year,marks,marker="*",color="red", s=50)
var = [50,100,150,200,250,300,350,400]
plt.subplot(1,2,2)
plt.scatter(year,var)
plt.show()
```
